#### Basic Commands

I have executed these commands in a Kali Linux terminal to familiarize myself with basic command-line operations. **note: These are only some of the basic commands I have tried.**

```bash
# To display the current working directory
pwd

# To list files and directories that are in the current directory.
ls 
# To move to another directory 
cd /path
# To copy a file
cp source_file destination_file
# To move or rename a file
mv old_file_name new_file_name
# To remove a file or a directory
rm file_name
rm -r directory_name
# To create a new directory
mkdir new_directory
```
____
#### Intermediate Commands
These commands are more complex and are used in file searching and text processing, etc. **note: These are only some of the commands I have tried.**
``` bash
# to search for text patterns within a file:
grep 'pattern' file_name
# to find files and directories in the file system.
find /path -name 'file_name'
# to display system processes
top 
# to show disk space usage
df -h 
# to display running processes
ps aux
```
____
#### Advanced Commands
These are advanced commands. They're used for networking, firewall configuration, and process management. 
```bash
# to configure a firewall 
sudo iptables -L
# to display network statistics 
netstat -tuln
# to analyze network packets
sudo tcdump -i eth0 
# to kill a process
kill PID
killall process_name
```

### 2. Install Kali Linux tools from github
Steps I followed to install tools for github: 
1. Clone the Github Repo:
   ``` bash
   git clone github.com/repo.git
    ```
2. Navigate to the Tool's Directory:
   ```bash
   cd repo_name
   ```
3. Install Dependencies
   ```bash
   sudo apt install dependency_name
   ```
4. Run the tool
   ```bash
   ./tool_name
   ```
